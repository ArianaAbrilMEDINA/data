# Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(10)
```

#1. Análisis de facturación



```python
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn.objects as so
```

```python
def ingresosTotales(x):
  total = 0
  for i in x['Precio']:
      total += i

  return total
ingresosPrimeraTienda = ingresosTotales(tienda)
ingresosSegundaTienda = ingresosTotales(tienda2)
ingresosTerceraTienda = ingresosTotales(tienda3)
ingresosCuartaTienda = ingresosTotales(tienda4)
print(f"Ingresos de la primera tienda: ${ingresosPrimeraTienda}")
print(f"Ingresos de la segunda tienda: ${ingresosSegundaTienda}")
print(f"Ingresos de la tercera tienda: ${ingresosTerceraTienda}")
print(f"Ingresos de la cuarta tienda: ${ingresosCuartaTienda}")
tiendas = [tienda, tienda2, tienda3, tienda4]
auxGraf = {}
for t in range(len(tiendas)):
     auxGraf[f"Tienda {t +1}"] = ingresosTotales(tiendas[t])
display(auxGraf)
```

```python
df_auxGraf = pd.DataFrame(list(auxGraf.items()), columns=['Tienda', 'Ingresos'])

(
    so.Plot(df_auxGraf, x="Tienda", y="Ingresos", color="Tienda")
    .add(so.Bars(width=0.7))
    .show()
)
```

# 2. Ventas por categoría

```python
categorias_unicas = tienda['Categoría del Producto'].unique()
categorias_unicas
```

```python
productos_por_categoria = tienda['Categoría del Producto'].value_counts()
productos_por_categoria
```

```python
from re import L
def calificarPorCategoria(x):
  l = {}
  for i in x['Categoría del Producto']:
    aux = x[x['Categoría del Producto']==i]
    if i not in l:
      l[i] = len(aux)
    else:
      l[i] += len(aux)
  return l

tiendas = [tienda, tienda2, tienda3, tienda4]
total_por_categoria = {}

for t in tiendas:
  l = calificarPorCategoria(t)
  for categoria, cant in l.items():
    if categoria not in total_por_categoria:
      total_por_categoria[categoria] = cant
    else:
      total_por_categoria[categoria] += cant

total_por_categoria
```

```python

df_auxGraf = pd.DataFrame(list(total_por_categoria.items()), columns=['Categoria', 'Cantidad'])

plt.figure(figsize=(18,5))
ax = plt.gca()

plot = (
    so.Plot(df_auxGraf, x="Categoria", y="Cantidad", color="Categoria")
    .add(so.Bars(width=0.7))
    .on(ax)
    .show()
)

ax.tick_params(axis='x', rotation=45)
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(tiendas):
  print(f"Ventas por categoría en Tienda {i+1}:")
  productos_por_categoria = t['Categoría del Producto'].value_counts()
  display(productos_por_categoria)
  print("-" * 30)
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

for i, t in enumerate(tiendas):
  plt.figure(figsize=(12, 6))
  sns.countplot(data=t, x='Categoría del Producto', order = t['Categoría del Producto'].value_counts().index, hue='Categoría del Producto', legend=False)
  plt.title(f'Distribución de Categorías de Productos en {nombres_tiendas[i]}')
  plt.xlabel('Categoría del Producto')
  plt.ylabel('Cantidad de Productos')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
```

# 3. Calificación promedio de la tienda

# 4. Productos más y menos vendidos

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

for i, t in enumerate(tiendas):
  print(f"Análisis de productos más y menos vendidos en {nombres_tiendas[i]}:")

  conteo_productos = t['Producto'].value_counts()

  productos_mas_vendidos = conteo_productos.head(5)
  print("\nProductos más vendidos:")
  display(productos_mas_vendidos)

  productos_menos_vendidos = conteo_productos.tail(5)
  print("\nProductos menos vendidos:")
  display(productos_menos_vendidos)
```

```python
for i, t in enumerate(tiendas):
  print(f"Análisis de productos más y menos vendidos en {nombres_tiendas[i]}:")
  plt.figure(figsize=(6, 4))
  sns.barplot(x=productos_mas_vendidos.index, y=productos_mas_vendidos.values, palette='viridis', hue=productos_mas_vendidos.index, legend=False)
  plt.title(f'Top 10 Productos Más Vendidos en {nombres_tiendas[i]}')
  plt.xlabel('Productos')
  plt.ylabel('Cantidad')
  plt.xticks(rotation=45)

  plt.show()

  plt.figure(figsize=(6, 4))
  sns.barplot(x=productos_menos_vendidos.index, y=productos_menos_vendidos.values, palette='magma', hue=productos_menos_vendidos.index, legend=False)
  plt.title(f'Top 10 Productos Menos Vendidos en {nombres_tiendas[i]}')
  plt.xlabel('Productos')
  plt.ylabel('Cantidad')
  plt.xticks(rotation=45)

  plt.show()

  print("-" * 50)
```

# 5. Envío promedio por tienda

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

for i, t in enumerate(tiendas):
  costo_envio = t['Costo de envío'].mean()
  print(f"Costo de envío promedio de {nombres_tiendas[i]}: ${costo_envio:.2f}")
```

```python
costos_envio_promedio = {}
for i, t in enumerate(tiendas):
  costo_envio = t['Costo de envío'].mean()
  costos_envio_promedio[nombres_tiendas[i]] = costo_envio

df_costos_envio = pd.DataFrame(list(costos_envio_promedio.items()), columns=['Tienda', 'Costo de Envío Promedio'])

(
    so.Plot(df_costos_envio, x="Tienda", y="Costo de Envío Promedio", color="Tienda")
    .add(so.Line(marker='o')) # Add marker to see points and connect them
    .label(title="Costo de Envío Promedio por Tienda")
    .show()
)
```

```python
# Concatenate all store dataframes
all_stores_df = pd.concat([tienda.assign(Tienda='Tienda 1'),
                           tienda2.assign(Tienda='Tienda 2'),
                           tienda3.assign(Tienda='Tienda 3'),
                           tienda4.assign(Tienda='Tienda 4')])

# Create a strip plot to show the distribution of shipping costs per store
plt.figure(figsize=(10, 6))
sns.violinplot(data=all_stores_df, x="Tienda", y="Costo de envío", hue="Tienda", legend=False)
plt.title("Distribución de Costos de Envío por Tienda")
plt.show()
```
